```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Загружаем изображение
image = cv2.imread('k.jpg')

# Преобразуем изображение из BGR в RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])
sharpened = cv2.filter2D(image, -1, kernel)
sharpened_2 = cv2.addWeighted(image, 1.5, blurred_image, -0.5, 0)

edges = cv2.Sobel(image, cv2.CV_64F, 1, 1, ksize=5)
edges = cv2.convertScaleAbs(edges)

combined = cv2.addWeighted(blurred_image, 0.5, edges, 0.5, 0)
combined_2 = cv2.addWeighted(combined, 0.5, sharpened, 0.5, 0)

# Создаем фигуру с двумя подграфиками
plt.figure(figsize=(40, 5))

# Первый подграфик для оригинального изображения
plt.subplot(2, 3, 1)
plt.imshow(image)
plt.axis('off')  # Отключаем оси

# Второй подграфик для размытого изображения
plt.subplot(2, 3, 2)
plt.imshow(blurred_image)
plt.axis('off')  # Отключаем оси

# Третий график для резкости
plt.subplot(2, 3, 3)
plt.imshow(sharpened)
plt.axis('off')

# Четвертый
plt.subplot(2, 3, 4)
plt.imshow(sharpened_2)
plt.title('Sharpened Image 2')
plt.axis('off')

#Пятый
plt.subplot(2, 3, 5)
plt.imshow(edges)
plt.title('edges')
plt.axis('off')

#Шестой
plt.subplot(2, 3, 6)
plt.imshow(combined)
plt.axis('off')

# Показываем изображения
plt.show()
```
